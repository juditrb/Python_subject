```python
def calculate_protein_ratio(filename, relative_threshold, absolute_threshold, residue):
    def get_frequencies(protein_sequence):
        total = len(protein_sequence)
        count = protein_sequence.count(residue)
        relative_freq = count / total if total > 0 else 0
        return count, relative_freq

    with open(filename, 'r') as file:
        content = file.read()

    # Parsing the FASTA file
    entries = content.split('>')
    #print(entries)
    proteins = [entry.split('\n', 1)[1].replace('\n', '') for entry in entries if entry]
    print(proteins)
    # Counting proteins that meet the criteria
    qualifying_proteins = 0
    for protein in proteins:
        absolute_freq, relative_freq = get_frequencies(protein)
        if absolute_freq >= absolute_threshold and relative_freq >= relative_threshold:
            qualifying_proteins += 1

    return qualifying_proteins / len(proteins) if proteins else 0

# Example usage
filename = 'example_fasta_file.fa'
relative_threshold = 0.1  # Example threshold
absolute_threshold = 5    # Example threshold
residue = 'A'             # Example residue
ratio = calculate_protein_ratio(filename, relative_threshold, absolute_threshold, residue)
print(f"Ratio: {ratio}")
```
