````python
def calculate_aminoacid_frequencies(fasta_filename, 
                                    subsequences_filename,
                                    number_of_repetitions,
                                    output_filename ):
    """
    Calculate the relative and absolute frequency of proteins in the file
    named fasta_filename containing each of the sub-sequences described in
    file the file named subsequences_filename at least n times. Save it in
    an output file ordered by the proportion value (descending order)
    """
    
    #1. START A COUNTER: SPECIFIC COUNTER FOR EACH SUBSEQUENCE
    
    frequencies = {} # Initialize a dictionary
    num_prots = 0
    
    #2. READ THE SUBSEQUENCES FILE
    
    with open(subsequences_filename) as f:
        for line in f:
            subsequence = line.strip() # Remove leading characters: new lines, spaces...
            if subsequence != "":
                frequencies[subsequence] = 0 # Start the counter for each subsequence
    
    #3. READ AND PROCESS THE FASTA FILE
    
    with open(fasta_filename) as f:
        
        sequence = ""
        
        for line in f:
            
            if line.startswith(">"):
                
                num_prots += 1

                for subsequence in frequencies:     
                    if sequence.count(subsequence) >= number_of_repetitions:
                        frequencies[subsequence] += 1
                
                sequence = ""
            
            else:
                sequence += line.strip()
 
    for subsequence in frequencies:     
        if sequence.count(subsequence) >= number_of_repetitions:
            frequencies[subsequence] += 1
                
    #4. SORT THE RESULTS

    ordered_list = sorted(frequencies,
                          key=lambda x: frequencies[x],
                          reverse=True)
    
    #5. WRITE THE OUTPUT FILE
    with open(output_filename, "w") as f:
        
        f.write("{}{}\n".format(
                "#Number of proteins:".ljust(30),
                str(num_prots).rjust(10)))
        
        f.write("{}{}\n".format(
            "#Number of subsequences:".ljust(30),
            str(len(frequencies)).rjust(10)))
                                
        f.write("#subsequence proportions:\n")
        
        for subsequence in ordered_list:
            
            f.write("{}{}{}\n".format(
                subsequence.ljust(10), 
                str(frequencies[subsequence]).rjust(10),
                ("%.4f"%(frequencies[subsequence]/num_prots)).rjust(20)))
                
        
````
